```python
toplam_female_count_by_borough = df.groupby('boroughname')['female_count'].sum()
print(toplam_female_count_by_borough)
```

```python
subset_df = df[['boroughname', 'female_count']]
toplam_female_count_by_borough = subset_df.groupby('boroughname')['female_count'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='boroughname', y='female_count', data=toplam_female_count_by_borough, palette='viridis')
plt.title('Toplam Female Count by Borough')
plt.xlabel('Borough Name')
plt.ylabel('Toplam Female Count')
plt.show()
```
